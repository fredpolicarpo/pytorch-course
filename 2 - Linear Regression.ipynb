# Pytorch Course
## Linear Regression in the PyTorch way

```python
import torch

device = torch.device("cuda:0") 

x_data = torch.Tensor([[1.], [2.], [3.]], device = device)
y_data = torch.Tensor([[2.], [4.], [6.]], device = device)
```

```python
class Model(torch.nn.Module):
    def __init__(self):
        """
        In the constructor we instantiate two nn.Linear module
        """
        super(Model, self).__init__()
        self.linear = torch.nn.Linear(1,1) # Dimension of input=1 and Dimension of output = 1
        
        
    def forward(self, x): 
        """
        In the forward function we accept a Tensor of input data and we must return
        a Tensor of output data. We can use Modules defined in the constructor as 
        well as arbitrary operators on Tensor
        """
        y_pred = self.linear(x)
        return y_pred       
    
```

```python
model = Model()
```

```python
# Construct our loss function
criterion = torch.nn.MSELoss(reduction='sum')
```

```python
# Optimizer
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)
```

```python
#Training: forward, loss, backward, step


#Training Loop
for epoch  in range(500):
    # Forward pass: Compute predicted y by passing x to the model
    y_pred = model(x_data)
    
    # Compute and print loss
    loss = criterion(y_pred, y_data)
    print(epoch, loss.item())
    
    # Zero gradients, perform a backward pass, and update the weights
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
# Testing Model

# After training
hour_tensor = torch.Tensor([[4]])
print("predicted (after training)", 4, model.forward(hour_tensor).item())
```
